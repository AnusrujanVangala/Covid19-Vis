```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("covid_19_india.csv")

# Step 2: Strip whitespace from column names
df.columns = df.columns.str.strip()

# Step 3: Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 4: Replace '-' with 0 and convert relevant columns to numeric
numeric_cols = ['ConfirmedIndianNational', 'ConfirmedForeignNational',
                'Cured', 'Deaths', 'Confirmed']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col].replace('-', 0), errors='coerce')

# Step 5: Fill missing values (optional)
df[numeric_cols] = df[numeric_cols].fillna(0)

# Step 6: Group by Date to create daily summary
daily_summary = df.groupby('Date').agg(
    new_cases=('Confirmed', 'sum'),
    daily_deaths=('Deaths', 'sum')
).reset_index()

# Step 7: Create cumulative totals
daily_summary['total_cases'] = daily_summary['new_cases'].cumsum()
daily_summary['total_deaths'] = daily_summary['daily_deaths'].cumsum()

# Step 8: Show result
print(daily_summary.head())
```

```python
import pandas as pd
from prophet import Prophet

# Sample data
df = pd.DataFrame({
    'ds': pd.date_range(start='2020-01-01', periods=100, freq='D'),
    'y': range(100)
})

# Initialize and fit model
model = Prophet()
model.fit(df)

# Forecast
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Plot
model.plot_components(forecast)
```
